## Задание на 4 КТ

В наборе данных org_all.xlsx представлены характеристики банков по показателям: уставной капитал, работающие активы, капитал. Все показатели в млрд. руб. 

1. Рассчитать средние показатели и показатели вариации по несгруппированным данным: среднее арифметическое, среднее геометрическое, среднее геометрическое, медиана, дисперию, среднее квадратическое отклонение, коэффициент вариации по показателям капитала, уставного капитала, работающих активов. Построить для них диаграмму "ящик с усами" Сделать выводы об однородности статистических совокупностей. Проверить нормальность распределения всех показателей.
2. Оценить коэффициенты ассиметрии и эксцесса по показателям капитала, уставного капитала, работающих активов.
3. Провести оценку связи между всеми признаками (Пирсона, Спирмена, Кендалла - в зависимсости от результатов теста на нормальность), построив соответствующие корреляционые диаграммы, проверить статистическую значимость всех корреляций. Провести оценку частных коэффициентов корреляции.

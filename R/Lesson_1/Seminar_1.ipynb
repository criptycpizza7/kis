## Знакомство со средой статистических вычислений R

**Назначение R, история возникновения**

R — язык программирования для статистической обработки данных и работы с графикой, а также свободная программная среда вычислений с открытым исходным кодом в рамках проекта GNU. Язык создавался как аналогичный языку S, разработанному в Bell Labs, и является его альтернативной реализацией. Изначально R был разработан сотрудниками статистического факультета Оклендского университета Россом Айхэкой (англ. Ross Ihaka) и Робертом Джентлменом (англ. Robert Gentleman) (первая буква их имён — R), на момент 2011 года язык и среда поддерживаются и развиваются организацией R Foundation. https://www.r-project.org/

R широко используется как статистическое программное обеспечение для анализа данных и фактически стал стандартом для статистических программ. R доступен под лицензией GNU GPL. 

 В октябре 2011 года корпорация Oracle выпустила аппаратно-программный комплекс Big Data Appliance — NoSQL-кластер серверов массово-параллельной обработки, с интегрированным программными средствами на основе языка R и Apache Hadoop, а в феврале 2012 года язык встроен в Oracle Database. Также массово-параллельный анализ средствами R поддержан в аппаратно-программных комплексах Netezza корпорации IBM.

Два-три раза в год выходит свободно-распространяемый информационный журнал R Journal. Он содержит информацию по статистической обработке данных и разработке, что может быть интересно как пользователям, так и разработчикам R. 
http://journal.r-project.org/index.html

Одна из самых популярных конференций, посвящённых языку — useR! (The R User Conference), проходит ежегодно, начиная с 2004 года, собирает специалистов в различных областях. https://www.r-project.org/conferences.html

Универсальная функция комбинирования аргументов в вектор: c(...)
x<-c(1,2,3)} - Переменной x присваивается вектор (1,2,3)
Доступ к элементам вектора: 

**x[1]; x[3:8]; x[c(3:7,9,10)]**

Доступ к элементам с логическим аргументом:

**x[x>4]**

Поэлементное сложение, умножение векторов (даже разной размерности): 

**a<-c(1:5); b<-c(3:9);c<-a+b,d<-a*b**


Сумма, произведение, среднее, медиана элементов вектора: 

**sum(a); prod(b); mean(a); median(b);**

**!=**  - не равно, **< , <= , > , >=**

**&** - и

**|** - или,

Загрузка данных в формате csv (по умолчанию разделитель - запятая): 

**read.csv('file',header=TRUE(FALSE),sep=',')**

Загрузка ранее сохранённых данных: 

**load('file')**

Сохранить все объекты: 

**save.image('file')**

Сохранить выбранные объекты: **save(x,y,z,'file')**

Печатает список всех объектов в рабочей области: **ls()**

Печатает детали по выбранному объекту: **str(имя объекта)**

Печатает свои аргументы: **print(x)**

Простая статистика по объекту: **summary(имя объекта)**

Удалить выбранный объект: **rm(имя объекта)**

Удалить все объектвы: **rm(list=ls())**

Некоторые полезные функции:

**abs,sign,sqrt,floor,ceiling,trunc,round,exp,log,log10,cos,sin,tan,acos,asin,atan**

**choose(n,k)=**$\frac{n!}{k!\cdot(n-k)!}$ - число сочетаний из n по k

**sample(x,size)** - случайная выборка элементов из вектора x без повторения

**sample(x,size,replace= TRUE)** - случайная выборка элементов из вектора x с повторениями

**runif(count, low, high)** - Генерирует count случайных вещественных чисел из диапазона low..high

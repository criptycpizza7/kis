# Задания к зачету по дисциплине

1. В командной оболочке bash выполнить скрипт, который создаст базу данных dot, создаст в ней таблицу из 20 столбцов. Например, 
> echo "create database dot; use dot; CREATE TABLE `my`(`a` INT PRIMARY KEY NOT NULL,`b` VARCHAR(60) NOT NULL);" |mysql -u juna -p
2. На языке Python, прочитав данные из файла data.csv, записать их в таблицу созданной базы данных dot.
3. С помощью atoti построить гиперкуб и рассчитать количество непустых значений в каждом столбце.
4. Развернуть кластерную инфраструктуру Redis, с двумя мастерами и тремя репликами на каждый.
5. С помощью Python прочитать данные из созданной базы данных dot, переписав их в созданную кластерную структуру Redis хранилища. Каждую строчку таблицы, исключая пустые значения, необходимо записать в Redis с помощью структуры списка, генерируя для каждой строки уникальный идентификатор.
6. С использованием Python рассчитать количество элементов каждого списка Redis. Полученный численный массив записать в Redis в виде структуры номер списка - значение.
7. На языке Python реализовать функцию, которая выбирает случайный список Redis, в котором выбирает случайный элемент списка, полученные таким образом слова конкатенируются в строку, которая возвращается в качестве ответа.
8. Развернуть кластерную инфраструктуру MongoDB из 2 кластеров по одному мастеру и одной реплике в каждом, с двумя конфигурационными серверами и двумя маршрутизаторами. 
9. С использованием функции, созданной в пункте 7, сгенерить 1000 строк. Каждую строку записать в MongoDB в виде отдельного документа, в котором каждое слово строки записывается под своим атрибутом. Например, строка "I am happy" превращается в документ {name1:"I", name2:"am", name3:"happy"}
10. Найти все документы, в которых в name1 содержится буква "a", а в name2 буква "m".

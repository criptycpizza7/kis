# Семинар 4. Контрольная работа 1

1. Подключиться к серверу MySQL, выполнить запрос к БД monitor_vuz на подсчет количества вузов (таблица sp_vuz) в разрезе регионов (таблица sp_region). Полученный результат запроса сохранить в виде структуры Python словарь, в словаре ключом должно быть название региона, значением - количество городов.

2. С использованием библиотеки socket языка Python создать сервер, который слушает порт 3000, сервер должен принимать строки от клиентов и сохранять их в двухсвязный список, реализованный в виде классов на языке Python.

3. Запустить Redis сервер на порту 3500, записать в хранилище Redis:
 -  словарь, созданный в пункте 1 в виде ключ - название региона, значение - строка из количества;
 -  словарь, созданный в пункте 1 в виде в виде ключ - uuid, значение - хеш из ключ-значение, где ключом выступает название региона, значением количество вузов в регионе. Создать 20 таких uuid, распределив хранение в них данных из словаря пункта 1;
 - двухсвязный список созданный в пункте 2 в виде списка Redis, сохранив весь список под одним идентификатром (ключом) uuid.
 
4. Развернуть кластер Redis из 12 компьютеров по 3 реплики на мастер. Записать в развернутый кластер данные, сохраненные на Redis-сервере в пункте 3. Нарисовать структуру созданного кластера. Смоделировать сбой в работе кластера, удалив 2 мастера, нарисовать структуру перестроенного кластера. Создать в MySQL новую БД с именем Redis, переписав в нее все данные из созданного кластера Redis, структуру таблиц выбрать на свое усмотрение.

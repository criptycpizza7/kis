# Семинар 3. Работа с данными

```python
#Подключаем библиотеки
import numpy as np
import pandas as pd

#Читаем данные с github.com/junaart/DataForR
data=pd.read_csv("https://raw.githubusercontent.com/junaart/DataForR/main/org_all.csv",delimiter=",")

#Отображаем данные
display(data)
```

```python
#Изменим название столбца
data=data.rename(columns={'Unnamed: 0': 'org'})
data
```

```python
#Получить только показатели
data[['p1','p2','p3','p4','p5','p6']]
```

```python
#Отберем все организации и их показатели из Дубны
data[data['city']=='Дубна']
```

```python
#Отберем все организации из Дубны,
data[data['city']=='Дубна']['org']
```

```python
#Отберем все организации из Дубны, у которых показатель p2>10
data[(data['city']=='Дубна') & (data['p2']>=10)][['org','p2']]
```

```python
#Построим Series из средних значений каждого показателя
s=pd.Series([data['p1'].mean(), data['p2'].mean(), data['p3'].mean(), data['p4'].mean(), data['p5'].mean(), data['p6'].mean()])
s
```

```python
#Получить только показатели и присвоим строкам название организаций
dd=data[['p1','p2','p3','p4','p5','p6']]
dd.index=data['org']
display(dd)
```

```python
#Построим круговую диаграмму по показателю p1
dd['p1'].plot.pie()
```

```python
#Построим точечную диаграмму по показателям p1, p2
dd[['p1','p2']].plot.scatter(x='p1',y='p2')
```

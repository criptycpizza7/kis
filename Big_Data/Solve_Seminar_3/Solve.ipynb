```python
#1. Загрузить фрейм данных 'demo26.xlsx' с сайта https://github.com/junaart/ForStudents/tree/master/Big_Data/Seminar_3
import pandas as pd
import numpy as np

data=pd.read_excel("https://github.com/junaart/ForStudents/blob/master/Big_Data/Seminar_3/demo26.xlsx?raw=true")
display(data)
```

```python
#2. Создать фрейм данных, транспонируя исходный фрейм данных, присвоив имена столбцов 'FIRST', 'SECOND', 'THIRD', 'FOURTH', 'FIFTH'.
dd=(data.T).iloc[0:,0:5]
dd.columns=['FIRST', 'SECOND', 'THIRD', 'FOURTH', 'FIFTH']
display(dd)
```

```python
#3. Отобрать года, для которых значение во втором столбце больше среднего значения этого столбца.
ddd=dd[1:]
display(ddd)
print(ddd.SECOND.mean())
s_3=ddd[ddd.SECOND>ddd.SECOND.mean()].index
s_3
```

```python
#Вывести на экран все строки фрейма за четные года.
L=list(ddd.index)
LL=pd.DataFrame(columns=ddd.columns)
for i in L:
    if i%2==0:
        LL.loc[i]=ddd.loc[i]

LL
```

```python
#Построить график по оси x - года, по оси y значения третьего столбца
ddd.THIRD.plot()
```

```python
#Добавить к фрейму новый столбец категориальных данных по принципу: 
#если значение третьего столбца в строке больше медианного значения 
#этого столбца и значение четвертого столбца этой строки больше медианного 
#значения этого столбца, то в новый столбец пишем 'good', в противном случае пишем 'bad'.
med_3=ddd.THIRD.median()
med_4=ddd.FOURTH.median()
ss=[]
for i in list(ddd.index):
    if (ddd.loc[i][2]>med_3) & (ddd.loc[i][3]>med_4):
        ddd.loc[i,'g_b']='good'
    else:
        ddd.loc[i,'g_b']='bad'
ddd
```

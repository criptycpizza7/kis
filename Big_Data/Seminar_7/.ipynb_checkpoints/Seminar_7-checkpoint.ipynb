4. Провести анализ статистической значимости коэффициентов корреляции между всеми количественными признаками датасета iris. 

```python
import pydataset as ds 
from math import *
df= ds.data('iris')
df.columns=['sl','sw','pl','pw','s']
df_sl = df['sl']
sl_Disp = df_sl.std()**2
df_setosa = df[df['s'] == 'setosa']['sl']
df_versicolor = df[df['s'] == 'versicolor']['sl']
df_virginica = df[df['s'] == 'virginica']['sl']
setosa_sl_Disp = df_setosa.std()**2
versicolor_sl_Disp = df_versicolor.std()**2
virginica_sl_Disp = df_virginica.std()**2
Disp_avg_in = (setosa_sl_Disp*df_setosa.count()+
               versicolor_sl_Disp*df_versicolor.count()+
               virginica_sl_Disp*df_virginica.count())/len(df.index)

setosa_sl_avg = df_setosa.mean()
versicolor_sl_avg = df_versicolor.mean()
virginica_sl_avg = df_virginica.mean()
Avg_general = df['sl'].mean()
Disp_between = ((setosa_sl_avg-Avg_general)**2*df_setosa.count()+
(versicolor_sl_avg-Avg_general)**2*df_versicolor.count()+
(virginica_sl_avg-Avg_general)**2*df_virginica.count())/df['sl'].count()
print(sl_Disp)
print(Disp_between+Disp_avg_in)
emp_corr_1 = sqrt(Disp_between/sl_Disp)
emp_corr_2 = sqrt(1-Disp_avg_in/sl_Disp)
print(emp_corr_1)
print(emp_corr_2)
```

3. Отобрать значимые коэффициенты корреляции в полученном датафрейме, визуализировать полученную корреляционную матрицу.

```python
from scipy.stats import t
def corr_s(arr,n,alpha):
    res=np.array(arr)
    for i in range(np.shape(arr)[0]):
        for j in range(np.shape(arr)[0]):
            if i!=j:
                tn=math.fabs(res[i,j])*math.sqrt(n-2)/math.sqrt(1-res[i,j]**2)
                tk=t.ppf(1 - alpha, n-2)
                if tn<tk:
                    res[i,j]=0
    res=pd.DataFrame(res)
    res.index=arr.index;res.columns=arr.columns
    return res

C_s=corr_s(C, 13,0.05)
C_s.style.background_gradient(cmap='viridis')
```

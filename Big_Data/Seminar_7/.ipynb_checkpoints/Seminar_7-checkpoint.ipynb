4. Провести анализ статистической значимости коэффициентов корреляции между всеми количественными признаками датасета iris. 

```python
import pydataset as ds 
from math import *
df= ds.data('iris')
df.columns=['sl','sw','pl','pw','s']
df_sl = df['sl']
sl_Disp = df_sl.std()**2
df_setosa = df[df['s'] == 'setosa']['sl']
df_versicolor = df[df['s'] == 'versicolor']['sl']
df_virginica = df[df['s'] == 'virginica']['sl']
setosa_sl_Disp = df_setosa.std()**2
versicolor_sl_Disp = df_versicolor.std()**2
virginica_sl_Disp = df_virginica.std()**2
Disp_avg_in = (setosa_sl_Disp*df_setosa.count()+
               versicolor_sl_Disp*df_versicolor.count()+
               virginica_sl_Disp*df_virginica.count())/len(df.index)

setosa_sl_avg = df_setosa.mean()
versicolor_sl_avg = df_versicolor.mean()
virginica_sl_avg = df_virginica.mean()
Avg_general = df['sl'].mean()
Disp_between = ((setosa_sl_avg-Avg_general)**2*df_setosa.count()+
(versicolor_sl_avg-Avg_general)**2*df_versicolor.count()+
(virginica_sl_avg-Avg_general)**2*df_virginica.count())/df['sl'].count()
print(sl_Disp)
print(Disp_between+Disp_avg_in)
emp_corr_1 = sqrt(Disp_between/sl_Disp)
emp_corr_2 = sqrt(1-Disp_avg_in/sl_Disp)
print(emp_corr_1)
print(emp_corr_2)
```

3. Отобрать значимые коэффициенты корреляции в полученном датафрейме, визуализировать полученную корреляционную матрицу.

```python
from scipy.stats import t
def corr_s(arr,n,alpha):
    res=np.array(arr)
    for i in range(np.shape(arr)[0]):
        for j in range(np.shape(arr)[0]):
            if i!=j:
                tn=math.fabs(res[i,j])*math.sqrt(n-2)/math.sqrt(1-res[i,j]**2)
                tk=t.ppf(1 - alpha, n-2)
                if tn<tk:
                    res[i,j]=0
    res=pd.DataFrame(res)
    res.index=arr.index;res.columns=arr.columns
    return res

C_s=corr_s(C, 13,0.05)
C_s.style.background_gradient(cmap='viridis')
```

4. Сформировать показатель суммарное количество публикаций в Scopus и Web Of Science, определить для него максимальный коэффициент детерминации по три переменных.

```python
import random
dd=data.iloc[:,1:]
dfCols = dd.columns[1:]
dfCols=dfCols[(dfCols!='P3_2') & (dfCols!='P3_3')]

det_corrs = []

for i in range(200):
    cols = random.choices(dfCols,k=3)
    df_rand = pd.DataFrame()
    
    df_rand['sum'] = dd['P3_2']+dd['P3_3']
    for col in cols:
        df_rand[col] = dd[col]

    corr_rand = df_rand.corr()

    det_corr = 1-np.linalg.det(corr_rand)/np.linalg.det(np.array(corr_rand)[1:,1:])
    det_corrs.append((det_corr,cols))

det_corrs=sorted(det_corrs,key=(lambda x: x[0]))
det_corrs = pd.Series(det_corrs)

for row in det_corrs:
    print('{0:6.3f} | {1}'.format(row[0],row[1]))
```

```python
import pandas as pd
iris = ds.data('iris')
MyData=pd.DataFrame({'sl':iris['Sepal.Length'],
                  'sw':iris['Sepal.Width'],})
dfr=MyData.rank()
```

```python
def Txy(x):
    from collections import Counter
    count=list((Counter(x)).values())
    res=0
    for i in count:
        res+=i**3-i
    return 1/2*res

n=len(dfr.index)
Tx=Txy(dfr['sl'])
Ty=Txy(dfr['sw'])
nn=(1/6)*(n**3-n)
```

```python
(nn-((dfr['sl']-dfr['sw'])**2).sum()-Tx-Ty)/((nn-2*Tx)*(nn-2*Ty))**(1/2)
```

```python
def MyKendall(dataFrame,columns):
    P=0
    Q=0

    df_rank = dataFrame[columns].rank()
    df_rank_sort_sl = df_rank.sort_values(by=columns[0])
    
    for i in range(df_rank_sort_sl.shape[0]):
        df_1 = df_rank_sort_sl.iloc[i+1:,1]
        P+= df_1[df_1 > df_rank_sort_sl.iloc[i,1]].count()
    
    for i in range(df_rank_sort_sl.shape[0]):
        df_1 = df_rank_sort_sl.iloc[i+1:,1]
        Q+= df_1[df_1 < df_rank_sort_sl.iloc[i,1]].count()
    
    S=P-Q
    n = df_rank_sort_sl.shape[0]
    
    return 2*S/(n*(n-1))

print(MyKendall(MyData[['sl','sw']],['sl','sw']))
```

```python
def corrKendall(dataFrame):
    res=np.zeros((dataFrame.shape[1],dataFrame.shape[1]))
    for i in range(dataFrame.shape[1]):
        for j in range(dataFrame.shape[1]):
            if(i==j):
                res[i,j] =1
            else:
                res[i,j]=MyKendall(dataFrame, [dataFrame.columns[i],dataFrame.columns[j]])
    res=pd.DataFrame(res)
    res.index=dataFrame.columns
    res.columns=dataFrame.columns
    return res
```

```python
import pydataset as ds 
df= ds.data('iris')
df.columns=['sl','sw','pl','pw','s']

CorrKendall_build = (df.loc[:,'sl':'pw']).corr('kendall')
display(CorrKendall_build)
display(corrKendall(df.loc[:,'sl':'pw']))
```
